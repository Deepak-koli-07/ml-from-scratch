#### What is an Embedding?

An embedding is a way to convert text into numbers (vectors) so a computer can understand and compare meaning.

Computers don’t understand words directly, so we convert each word, sentence, or document into a vector.

With embeddings, we can do:
- similarity search
- classification
- clustering
- retrieval (R in RAG)
- memory
- recommendations

------------------------------------------------------------

#### Why do we need embeddings?

Computers only understand numbers, not English or Hindi.

If two texts have similar meaning → their vectors should be close.
If meaning is different → vectors should be far apart.

This lets us measure similarity using math (like cosine similarity).

------------------------------------------------------------

#### Custom Embedding Using Bag-of-Words (BoW)

We built our own small embedding model.

Step 1: Take all documents
Doc1: "apple banana apple"
Doc2: "banana orange"
Doc3: "orange banana apple apple"

Step 2: Extract all unique words → Vocabulary
Vocabulary = ["apple", "banana", "orange"]

This is the feature space.

------------------------------------------------------------

#### Step 3: Assign an index to each word

Word     Index
apple      0
banana     1
orange     2

Why indexing?
Because a vector needs fixed positions:
- index 0 always means “apple”
- index 1 always means “banana”
- index 2 always means “orange”

Indexing = mapping words → vector positions.

Example vector: [2, 1, 0]
Meaning:
- apple  = 2
- banana = 1
- orange = 0

------------------------------------------------------------

#### Step 4: Convert each document into a vector

Doc1: "apple banana apple"
apple: 2, banana: 1, orange: 0
Vector → [2, 1, 0]

Doc2: "banana orange"
apple: 0, banana: 1, orange: 1
Vector → [0, 1, 1]

Doc3: "orange banana apple apple"
apple: 2, banana: 1, orange: 1
Vector → [2, 1, 1]

------------------------------------------------------------

#### Document–Term Matrix (BoW Matrix)

         apple  banana  orange
Doc1       2       1       0
Doc2       0       1       1
Doc3       2       1       1

Python-style matrix:
[
  [2, 1, 0],
  [0, 1, 1],
  [2, 1, 1]
]

This matrix is your custom embedding model.


```python
import math

# 1) Get unique words across all docs (vocabulary)
def get_vocab(*docs):
    combined = " ".join(docs).replace(",", "")
    words = combined.split()
    unique = sorted(set(words))  # sorted for fixed order
    return unique  # list of words, index = column


# 2) Convert one document to a vector using vocab
def doc_to_vec(doc, vocab):
    words = doc.replace(",", "").split()
    vec = []
    for w in vocab:
        vec.append(words.count(w))
    return vec


# 3) Build document-term matrix for many docs
def build_doc_term_matrix(docs):
    vocab = get_vocab(*docs)
    matrix = []
    for doc in docs:
        vec = doc_to_vec(doc, vocab)
        matrix.append(vec)
    return vocab, matrix


# 4) Pretty print matrix like you showed
def print_matrix(vocab, matrix):
    # header row
    header = "       " + " ".join(f"{w:7}" for w in vocab)
    print(header)
    
    # each document row
    for i, vec in enumerate(matrix, 1):
        row_name = f"Doc{i}"
        row_values = " ".join(f"{v:7}" for v in vec)
        print(f"{row_name:<7}{row_values}")




import math

def cosine_similarity(a, b):
    """
    Compute cosine similarity between two vectors a and b.
    Both a and b must be lists of numbers of the same length.
    """

    # dot product
    dot = sum(x * y for x, y in zip(a, b))

    # vector magnitudes (norms)
    norm_a = math.sqrt(sum(x * x for x in a))
    norm_b = math.sqrt(sum(y * y for y in b))

    # avoid division by zero
    if norm_a == 0 or norm_b == 0:
        return 0.0

    # cosine similarity formula
    return dot / (norm_a * norm_b)
```
