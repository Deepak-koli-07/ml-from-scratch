# Regression Metrics

#### We measure how far predictions are from actual values.

```python
import pandas as pd
```

```python
y = [10, 20, 30, 40]
y_hat = [12,18,25,42]
df = pd.DataFrame({'y' : y, 
                   'y_hat' : y_hat})
df.head()
```

## MAE - Mean Absolute Error
- On Average, how much was the model wrong 

```python
df['mod_y_y_hat'] = abs(df['y'] - df['y_hat'])
df.head()
```

```python
mae = (df['mod_y_y_hat'].sum())/(len(df))
print(f'prediction is approximately actual ± {mae}')
```

MAE means Mean Absolute Error — it ignores the direction of the error.

If MAE = 2.75, it means:

✔ Sometimes your model predicts 2.75 higher  
✔ Sometimes your model predicts 2.75 lower  
✔ The average mistake size is 2.75

In math form:

MAE = 2.75  ⇒  Prediction error ≈ ±2.75

So your model is off by about 2.75 units on average.


## MSE — Mean Squared Error

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$


```python
df.head()
```

```python
mse = sum([i*i for i in df['mod_y_y_hat']])/len(df)
mse 
```

What does it mean when MSE is high 


1. Your model is too simple → predictions are far from actual.

```python
print('Lets just say')
print(f'If my actual y is {[i for i in df['y']]} and my model predictions are {[25 for i in range(4)]}')
print(f'sum of error is {sum([(i - 25)**2 for i in df['y']])} and the mse is {sum([(i - 25)**2 for i in df['y']])/len(df['y'])}')
print('This is the case of High MSE because MODEL IS WRONG (Underfitting)')
```

2. Model is good, but one bad data point destroys MSE.

```python
y = [10, 20, 30, 200]
y_pred = [12, 22, 28 ,50]
error = [a - b for a,b in zip(y, y_pred)]
mse = sum([i*i for i in error])/len(y)
print(f'NOT model’s fault — last point is too far.One outlier added 22500 to MSE which is ------>.{mse}')

```

| Cause            | Why MSE is high?                     | Blame                  |
|------------------|----------------------------------------|------------------------|
| Underfitting     | Model is too simple                    | Model                  |
| Outliers         | One big error gets squared (explodes)  | Data                   |
| Missing features | Model is blind to important patterns   | Inputs / Preprocessing |

#### Why RMSE Is Important (Concise Points)

1. Same units as target
- MSE is in squared units (₹², kg², etc.)
- RMSE = sqrt(MSE) → converts error back to original units
- Example: RMSE = 3000 → model is off by ~₹3000

2. Easy to judge error size
- Compare RMSE with typical value of y
- Small RMSE vs y → good model
- Large RMSE vs y → bad model

3. Punishes big errors strongly
- Comes from squaring errors
- One large mistake increases RMSE a lot
- Helps detect outliers and unstable predictions

4. Standard metric everywhere
- Used in Kaggle, industry ML models, research
- More intuitive than MSE

5. Easy to compare models
- Smaller RMSE = better model
- Example:
  Model A: 13
  Model B: 18
  Model C: 10
  → C is best

6. RMSE vs MAE gives insights
- RMSE >> MAE → big errors/outliers exist
- RMSE ≈ MAE → errors are uniform

7. Helps detect overfitting
- Train RMSE << Test RMSE → overfitting
- RMSE clearly shows the gap


#### MAE vs MSE vs RMSE — Side-by-Side Comparison

| Metric | Meaning                 | Punishes Big Errors? | Easy to Understand? | Sensitive to Outliers? |
|--------|--------------------------|------------------------|-----------------------|--------------------------|
| MAE    | Average absolute error   | ❌ No                  | ✔ Very easy           | Low                      |
| MSE    | Average squared error    | ✔ Yes (squares)       | ❌ Hard (units squared)| High                     |
| RMSE   | Square root of MSE       | ✔ Yes (less than MSE) | ✔ Easy (same units)   | Medium                   |
